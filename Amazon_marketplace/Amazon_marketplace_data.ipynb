# Приведение столбцов к стилю camel_case

```python
# Функция для преобразования строки в camel case
def to_camel_case(s):
    # Разбиваем строку на слова по '_'
    words = s.split('_')
    # Превращаем первое слово в нижний регистр, остальные - в верхний регистр
    return words[0].lower() + ''.join(word.capitalize() for word in words[1:])

# Применяем функцию ко всем столбцам DataFrame
df.columns = [to_camel_case(col) for col in df.columns]

# Выводим обновленный DataFrame для проверки
df.head()
```

### По регионам

```python
import matplotlib.pyplot as plt

# 1. Подсчитаем количество уникальных пользователей по регионам
region_counts = df.groupby('region')['customer id'].nunique()

# 2. Отсортируем регионы по убыванию количества пользователей
sorted_region_counts = region_counts.sort_values(ascending=False)

# 3. Построим столбчатую диаграмму
plt.figure(figsize=(10, 6))  # Устанавливаем размер графика (ширина 10, высота 6 дюймов)
sorted_region_counts.plot(kind='bar', color='blue')  # Рисуем столбчатую диаграмму синего цвета

# 4. Добавим заголовок и подписи к осям
plt.title('Количество пользователей по регионам')  # Заголовок графика
plt.xlabel('Регион')  # Подпись для оси X
plt.ylabel('Количество пользователей')  # Подпись для оси Y

# 5. Уберем сетку с графика
plt.grid(False)

# 6. Повернем метки на оси X для лучшей читаемости
plt.xticks(rotation=45)  # Поворачиваем метки на 45 градусов

# 7. Добавим подписи над столбиками
for index, value in enumerate(sorted_region_counts):  # Проходим по индексам и значениям
    plt.text(index, value + 0.5, str(value), ha='center')  # Добавляем текст над столбиком

# 8. Покажем график
plt.tight_layout()
plt.show()
```

## Динамика кол-ва заказов по месяцам

```python
# Осуществляем агрегацию данных по месяцам
# Создаем новый DataFrame, который будет содержать количество заказов по месяцам
monthly_orders = df.resample('M', on='order date').size().reset_index(name='order count')

# Строим график
plt.figure(figsize=(12, 6))  # Определяем размер графика
plt.plot(monthly_orders['order date'], monthly_orders['order count'], color='blue', marker='o')  # Строим график

# Подписываем точки на графике
for i, count in enumerate(monthly_orders['order count']):
    plt.text(monthly_orders['order date'].iloc[i], count, str(count), fontsize=9, ha='center', va='bottom')  # Подписи над точками

# Настраиваем график
plt.title('Динамика количества заказов по месяцам')  # Заголовок графика
plt.xlabel('Дата заказа (по месяцам)')  # Метка оси X
plt.ylabel('Количество заказов')  # Метка оси Y
plt.xticks(rotation=45)  # Поворачиваем подписи по оси X для удобства чтения
plt.grid(False)  # Убираем сетку

# Показать график
plt.tight_layout()  # Подгоняем график для лучшего отображения
plt.show()
```

Первый пик заказов приходится на апрель 2020г, 304 заказа.
С мая по сентябрь спад, цифры держатся в диапазоне от 229 до 251 заказа в месяц.
Два следующих пика приходятся на октябрь (385 заказаов) и декабрь (375 заказаов) 2020г.
В ноябре 2020 большой спад до 272 заказов. В январе количество заказаов также снижается.


## Динамика продаж по сумме по месяцам

```python
# Группируем данные по месяцам и суммируем продажи
monthly_sales = df.resample('M', on='order date')['sales'].sum().reset_index()

# Переименуем колонки для удобства
monthly_sales.columns = ['order_date', 'total_sales']

# Строим график
plt.figure(figsize=(12, 6))  # Устанавливаем размер графика
plt.plot(monthly_sales['order_date'], monthly_sales['total_sales'], marker='o', color='blue')  # Строим линию графика с точками

# Добавляем подписи для каждой точки
for i, row in monthly_sales.iterrows():
    plt.text(row['order_date'], row['total_sales'], f'{row["total_sales"]:.2f}',
             ha='center', va='bottom', fontsize=10)

# Настраиваем заголовок и метки осей
plt.title('Динамика продаж по сумме (Sales) по месяцам', fontsize=16)
plt.xlabel('Месяц', fontsize=14)
plt.ylabel('Сумма продаж', fontsize=14)

# Убираем сетку с графика
plt.grid(False)

# Поворачиваем метки по оси X для лучшей читаемости
plt.xticks(rotation=45)

# Показываем график
plt.tight_layout()  # Подгоняем элементы графика для лучшего отображения
plt.show()
```

В начале года самым прибыльным был апрель (74774 ед), в мае наблюдается самый большой спад за год: сумма продаж 29072 ед.
К сентябрю сумма продаж чуть превышает апрельский пик (75675 ед), в ноябре небольшой спад (65501 ед), а в декабре максимальный пик: 89306 ед). После чего в январе спад до 56969 ед.

Апрельский пик по сумме продаж совпадает с пиком по количеству заказов.


## Распределение продаж по категориям и подкатегориям товаров

```python
# Создание сводной таблицы для суммирования продаж по категориям
category_sales = df.pivot_table(values='sales', index='category', aggfunc='sum').reset_index()

# Сортировка сводной таблицы по убыванию суммарных продаж
category_sales = category_sales.sort_values(by='sales', ascending=False)

# Создание сводной таблицы для суммирования продаж по подкатегориям
subcategory_sales = df.pivot_table(values='sales', index='sub-category', aggfunc='sum').reset_index()

# Сортировка сводной таблицы по убыванию суммарных продаж
subcategory_sales = subcategory_sales.sort_values(by='sales', ascending=False)

# Визуализация суммарных продаж по категориям
plt.figure(figsize=(10, 6))  # Установка размера графика
bars = plt.bar(category_sales['category'], category_sales['sales'], color='blue')  # Столбчатая диаграмма
plt.title('Суммарные продажи по категориям')  # Заголовок графика
plt.xlabel('Категории')  # Подпись оси X
plt.ylabel('Суммарные продажи')  # Подпись оси Y
plt.xticks(rotation=45)  # Поворот меток по оси X для удобства чтения
plt.grid(False)  # Убрать сетку
plt.tight_layout()  # Автоматическая регулировка размещения элементов

# Добавление подписей значений на столбиках
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

plt.show()  # Показать график

# Визуализация суммарных продаж по подкатегориям
plt.figure(figsize=(12, 6))  # Установка размера графика
bars = plt.bar(subcategory_sales['sub-category'], subcategory_sales['sales'], color='blue')  # Столбчатая диаграмма
plt.title('Суммарные продажи по подкатегориям')  # Заголовок графика
plt.xlabel('Подкатегории')  # Подпись оси X
plt.ylabel('Суммарные продажи')  # Подписьоси Y
plt.xticks(rotation=45)  # Поворот меток по оси X для удобства чтения
plt.grid(False)  # Убрать сетку
plt.tight_layout()  # Автоматическая регулировка размещения элементов

# Добавление подписей значений на столбиках
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval), ha='center', va='bottom')

plt.show()  # Показать график
```

Самая популярная категория товаров Технология, а подкатегория - Телефоны.
На втором месте с небольшим отставанием категория Товары для офиса и подкатегория Стулья.
В категории Мебель сумма продаж ниже двух других. Меньше всего продаж в подкатегориях: Этикетки, Конверты, Крепежи - менее 5000.

```python
# Фильтруем данные для штата California
california_data = df[df['state'] == 'California']

# Группируем данные по 'state' и 'category', считаем количество уникальных заказов
# Здесь 'order id' уникален для каждого заказа
grouped_data = california_data.groupby(['state', 'category']).agg(order_count=('order id', 'nunique')).reset_index()

# Сортируем данные по количеству заказов в порядке убывания
sorted_data = grouped_data.sort_values(by='order_count', ascending=False)

# Строим график
plt.figure(figsize=(10, 6))
bars = plt.bar(sorted_data['category'], sorted_data['order_count'], color='blue')

# Настраиваем график
plt.title('Распределение категорий по заказам в штате California', fontsize=16)
plt.xlabel('Категория', fontsize=14)
plt.ylabel('Количество заказов', fontsize=14)
plt.xticks(rotation=45)  # Поворачиваем метки по оси X для лучшей читаемости
plt.grid(False)  # Убираем сетку

# Подписываем значения на графике
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom')  # Подписываем значение над баром

# Показываем график
plt.tight_layout()  # Упаковываем график
plt.show()
```

## Динамика продаж внутри категорий

```python
# Извлекаем год и месяц из даты для агрегации
df['year_month'] = df['order date'].dt.to_period('M')

# Группируем данные по год/месяц и категории, подсчитываем сумму продаж
sales_trend = df.groupby(['year_month', 'category'])['sales'].sum().reset_index()

# Убираем временной период в строковом формате для удобства визуализации
sales_trend['year_month'] = sales_trend['year_month'].dt.strftime('%Y-%m')

# Настраиваем график
plt.figure(figsize=(12, 6))

# Используем seaborn для создания линейного графика
sns.lineplot(data=sales_trend, x='year_month', y='sales', hue='category', palette='Blues_d', marker='o')

# Подписываем значения на графике
for line in plt.gca().lines:
    y_data = line.get_ydata()
    x_data = line.get_xdata()
    for x, y in zip(x_data, y_data):
        plt.text(x, y, f'{y:.0f}', fontsize=9, ha='right')

# Настраиваем заголовок и метки осей
plt.title('Динамика продаж по категориям')
plt.xlabel('Дата (Год-Месяц)')
plt.ylabel('Сумма продаж')
plt.xticks(rotation=45)  # Поворачиваем метки по оси X для лучшей читаемости
plt.grid(False)  # Убираем сетку
plt.legend(title='Категория')  # Добавляем легенду
plt.tight_layout()  # Подгоняем элементы графика
plt.show()
```

Самой прибыльной категорией является Технологии: максимальный пик продаж приходится на март (40701), также есть небольшие взлеты в июле (26054) и в ноябре (32539). Провалы в этой категории наблюдаются в июне (15611) и в декабре (13671).

В общем по стране категория Офисные принадлежности стоит на втором месте по прибыльности: аик продаж в этой категории приходится на август (36077), а самый провльный месяц - апрель (9896).

В категории Мебель пик продаж приходится на ноябрь (29818), максимальный спад в мае (10253).

В феврале, апреле, октябре и декабре наблюдаются спады во всех трех категориях товаров.А в ноябре идет рост продаж во всех трех категориях.


Дтнамика продаж по подкатегоряим внутри категории.

```python
# Создаем новый столбец 'year_month', чтобы группировать данные по месяцам
df['year_month'] = df['order date'].dt.to_period('M')

# Группируем данные по категории и подкатегории, суммируя продажи
sales_data = df.groupby(['category', 'sub-category', 'year_month'])['sales'].sum().reset_index()

# Получаем уникальные категории для построения отдельных графиков
categories = sales_data['category'].unique()

# Строим график для каждой категории
for category in categories:
    # Фильтруем данные по текущей категории
    category_data = sales_data[sales_data['category'] == category]

    # Создаем сводную таблицу для удобства визуализации
    pivot_table = category_data.pivot(index='year_month', columns='sub-category', values='sales')

    # Создаем график
    plt.figure(figsize=(12, 6))
    pivot_table.plot(kind='line', marker='o')  # Строим линейный график с маркерами

    # Добавляем заголовок и метки осей
    plt.title(f'Динамика продаж по подкатегориям в категории: {category}')
    plt.xlabel('Дата (год-месяц)')
    plt.ylabel('Общие продажи')
    plt.xticks(rotation=45)  # Поворачиваем метки по оси X для удобочитаемости

    # Убираем сетку
    plt.grid(False)

    # Показываем легенду
    plt.legend(title='Подкатегории')

    # Показываем график
    plt.tight_layout()  # Упорядочиваем график
    plt.show()  # Отображаем график
```

```python
# 2. Создаем новый столбец с месяцем и годом для группировки
df['year_month'] = df['order date'].dt.to_period('M')

# 3. Фильтруем данные для подкатегории 'Copiers'
copiers_data = df[df['sub-category'] == 'Copiers']

# 4. Группируем данные по месяцам, суммируем продажи
monthly_sales_copiers = copiers_data.groupby('year_month')['sales'].sum().reset_index()

# 5. Строим линейный график
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales_copiers['year_month'].astype(str), monthly_sales_copiers['sales'], marker='o')

# 6. Настраиваем график
plt.title('Динамика продаж по подкатегории "Copiers" по месяцам', fontsize=16)
plt.xlabel('Месяц', fontsize=14)
plt.ylabel('Общие продажи', fontsize=14)
plt.xticks(rotation=45)  # Поворачиваем метки по оси X для удобства чтения

# 7. Убираем сетку
plt.grid(False)

# 8. Показываем график
plt.tight_layout()  # Настраиваем макет
plt.show()
```

В категории мебель рост продаж наблюдается осенью, преобладают категории Стулья и Столы.
В категории офисные принадлежности пик продаж приходится на август, лидирует подкатегория Binders.
В категории Технологии в Марте пик продажи Копировальной техники, во второй половине года хорошие продажи Телефонов. Пик продажи компьютеров приходится на апрель.
